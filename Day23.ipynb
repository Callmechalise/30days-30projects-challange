```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as ynf
```

```python
stocks=["SRBL","MBL"]
data=ynf.download(stocks,start="2020-01-01",end="2025-01-01")["Close"]
data.head()
```

```python
data.describe()
```

```python
plt.figure(figsize=(12, 6))

for stock in stocks:
    plt.plot(data.index, data[stock], label=stock)
plt.legend()
plt.title("Price Trend for MBL and SBL")
plt.xlabel("Year")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```
